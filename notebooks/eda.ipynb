```python
import pandas as pd 
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import data 
transactions = pd.read_csv("../data/bank_transactions.csv")

transactions.head(15)
```

```python
#filter by nameorig
id_counts = transactions['nameOrig'].value_counts()
ids_to_keep = id_counts[id_counts >= 2].index #collects all the indexes where id is greater than five
```

```python
filtered_df = transactions[transactions['nameOrig'].isin(ids_to_keep)]
```

```python
filtered_df.sort_values(by='nameOrig').head(20)
```

```python
# TODO: view the shape of your data

transactions.shape
```

```python
# TODO: Begin your EDA
# histplot for amount and isFraud
# using log_scale to more easily view the spread of the data
```

```python
# saving fraud in a separate variable and not fraud in another to more clearly see trends of behavior
fraud = transactions[transactions["isFraud"] == 1]
not_fraud = transactions[transactions["isFraud"] == 0]
```

```python
# checking how many rows are present in fraud
fraud.shape
```

```python
not_fraud.shape
```

```python
# Histplot for "amount", observing the difference in non-fraud and fraud

sns.histplot(data=fraud, x="amount", color='red', label="Fraud", bins=50, log_scale=True)
plt.title("Transactions Amounts by Fraud")
plt.show()

sns.histplot(data=not_fraud, x="amount", color='blue', bins=50, log_scale=(True, False))
plt.title("Transactions Amounts by Non-Fraud")
plt.show()
```

```python
# the mean for amounts in Fraud vs Not Fraud
print(f"The average amount in Fraud transactions is {fraud['amount'].mean().round(2)}")
print(f"The average amount in Non-Fraud transactions is {not_fraud['amount'].mean().round(2)}")
```

```python
#histogram for oldbalanceOrg and isFraud

sns.histplot(data=fraud, x="oldbalanceOrg", color="red", bins=50, log_scale=True)
plt.title("Distribution of Old Balance Origin by Fraud")
plt.show()

sns.histplot(data=not_fraud, x="oldbalanceOrg", color="blue", bins=50, log_scale=True)
plt.title("Distribution of Old Balance Origin by Non-Fraud")
plt.show()
```

```python
# average of old balance origin
print(f"The average old balance org in Fraud transactions is {fraud['oldbalanceOrg'].mean().round(2)}")
print(f"The average old balance org in Non-Fraud transactions is {not_fraud['oldbalanceOrg'].mean().round(2)}")
```

```python
#histogram for newbalanceOrg

sns.histplot(data=fraud, x="newbalanceOrig", color="red", bins=50, log_scale=True)
plt.title("Distribution of New Balance Origin by Fraud")
plt.show()

sns.histplot(data=not_fraud, x="newbalanceOrig", color="blue", bins=50, log_scale=True)
plt.title("Distribution of New Balance Origin by Not Fraud")
plt.show()
```

```python
#histogram for oldbalanceDest

sns.histplot(data=fraud, x="oldbalanceDest", color="red", bins=50, log_scale=True)
plt.title("Distribution of Old Balance Destination by Fraud")
plt.show()

sns.histplot(data=not_fraud, x="oldbalanceDest", color="blue", bins=50, log_scale=True)
plt.title("Distribution of Old Balance Destination by Not Fraud")
plt.show()
```

```python
print(f"The average old balance destination in Fraud transactions is {fraud['oldbalanceDest'].mean().round(2)}")
print(f"The average old balance destination in Non-Fraud transactions is {not_fraud['oldbalanceDest'].mean().round(2)}")
```

```python
#histogram for newbalanceDest and isFraud

sns.histplot(data=fraud, x="newbalanceDest", color="red", bins=50, log_scale=True)
plt.title("Distribution of New Balance Destination by Fraud")
plt.show()

sns.histplot(data=not_fraud, x="newbalanceDest", color="blue", bins=50, log_scale=True)
plt.title("Distribution of New Balance Destination by Not Fraud")
plt.show()
```

```python
print(f"The average new balance origin in Fraud transactions is {fraud['newbalanceOrig'].mean().round(2)}")
print(f"The average new balance origin in Non-Fraud transactions is {not_fraud['newbalanceOrig'].mean().round(2)}")
```
